```python
def enviar_certificado_jpg(nome, destinatario, caminho_imagem):
    msg = MIMEMultipart()

    corpo_email = f"""Olá {nome.split()[0]}, boa tarde! Tudo bem?

Meu nome é Enzo, sou diretor de projetos da Liga MackEmpreende. Em nome da liga, gostaria de te agradecer pela presença na Semana e espero que tenha contribuído para o seu desenvolvimento!

Além disso, estou te enviando, em anexo, seu certificado de horas complementares com um manual de submissão dele no TIA."""

    password = 'jkqpfvnjmdmauvto'

    # Infos email
    msg['Subject'] = "Certificado de Horas Complementares - MackEmpreende"
    msg['From'] = 'enzoribeiro.2005@gmail.com'
    msg['To'] = destinatario
    msg.attach(MIMEText(corpo_email, 'plain'))
    msg.add_header('Content-Type', 'text/html')

    # Anexar imagem
    img_data = open(caminho_imagem, 'rb').read()
    image = MIMEImage(img_data, name=os.path.basename(caminho_imagem))
    msg.attach(image)

    # Anexa o arquivo PDF
    arquivo_anexo = '/content/Manual de Submissão de Horas Complementares - MackEmpreende.pdf'

    with open(arquivo_anexo, 'rb') as anexo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(anexo.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', 'attachment', filename = "Manual de Submissão de Horas Complementares.pdf")
        msg.attach(part)

    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()

    # Login Credentials for sending the mail
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    print('Email enviado')
```

```python
def enviar_certificado_pdf(nome, destinatario, caminho_certificado):
    msg = MIMEMultipart()

    corpo_email = f"""Olá {nome.split()[0]}, boa tarde! Tudo bem?

Meu nome é Enzo, sou diretor de projetos da Liga MackEmpreende. Em nome da liga, gostaria de te agradecer pela presença na Semana e espero que tenha contribuído para o seu desenvolvimento!

Além disso, estou te enviando, em anexo, seu certificado de horas complementares com um manual de submissão dele no TIA."""

    password = 'jkqpfvnjmdmauvto'

    # Infos email
    msg['Subject'] = "Certificado de Horas Complementares - MackEmpreende"
    msg['From'] = 'enzoribeiro.2005@gmail.com'
    msg['To'] = destinatario
    msg.attach(MIMEText(corpo_email, 'plain'))
    msg.add_header('Content-Type', 'text/html')

    # Anexa o manual em PDF
    arquivo_anexo = '/content/Manual de Submissão de Horas Complementares - MackEmpreende.pdf'

    with open(arquivo_anexo, 'rb') as anexo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(anexo.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', 'attachment', filename = "Manual de Submissão de Horas Complementares.pdf")
        msg.attach(part)


    # Anexa o certificado em PDF
    with open(caminho_certificado, 'rb') as anexo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload(anexo.read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', 'attachment', filename = "Certificado Semana de Empreendedorismo")
        msg.attach(part)

    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()

    # Login Credentials for sending the mail
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    print('Email enviado')
```
